#Normalization

```python
features=['MDVP:Fo(Hz)', 'MDVP:Fhi(Hz)', 'MDVP:Flo(Hz)', 'MDVP:Jitter(%)',
       'MDVP:Jitter(Abs)', 'MDVP:RAP', 'MDVP:PPQ', 'Jitter:DDP',
       'MDVP:Shimmer', 'MDVP:Shimmer(dB)', 'Shimmer:APQ3', 'Shimmer:APQ5',
       'MDVP:APQ', 'Shimmer:DDA', 'NHR', 'HNR','RPDE', 'DFA',
       'spread1', 'spread2', 'D2', 'PPE']
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
parkinsons_df[features]=scaler.fit_transform(parkinsons_df[features])
```

```python
parkinsons_df.head()
```

```python
X=parkinsons_df[features]
y=parkinsons_df['status']
```

#Train_Test_split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.2,random_state=4)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

#PCA

```python
from sklearn.decomposition import PCA
pca=PCA(.95)
X_train=pca.fit_transform(X_train)
X_test=pca.transform(X_test)
```

```python
pca.explained_variance_ratio_
```

#Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(X_train,y_train)
```

```python
model.score(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
y_pred=model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
print(cm)
# y_test.shape,y_pred.shape
```

#KNN

```python
from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier()
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
y_pred
```

```python
model.score(X_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
print(cm)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model_rf=RandomForestClassifier(n_estimators=10,min_samples_split=2)
model_rf.fit(X_train,y_train)
model_rf.score(X_test,y_test)
```

```python
from sklearn.ensemble import AdaBoostClassifier
model_ad=AdaBoostClassifier(n_estimators=30,learning_rate=1)
```

```python
model_ad.fit(X_train,y_train)
model_ad.score(X_test,y_test)
```
